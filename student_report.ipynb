```python
print('***** Report Card *****')
name=input("\nStudent Name:")
math=float(input('Math Score:'))
english=float(input('English Score:'))
science=float(input('Science Score:'))
average=(math+english+science)/3
print(f'\n'+'='*50, '\n              Student Grade Report','\n'+'='*50)
print(f'\nStudent Name: {name:^30}')
print('\n'+'-'*50)
print(f'\nMath   :       {math:^29.1f}')
print(f'English:      {english:^30.1f}')
print(f'Science:      {science:^30.1f}')
print('\n'+'-'*50)
print(f'\nAverage:      {average:^30.1f}')
print('\n'+'-'*50)
if average>=90:
  grade='A'
elif average>=80:
  grade='B'
elif average>=70:
  grade='C'
elif average>=50:
  grade='D'
else:
  grade='F'
print(f'\nGrade  :       {grade:^30}')
print('\n'+'='*50)
```
