```python
print('=== Personal Info ===')

full_name=input('Enter your Full Name:')                                #Input Personal Information
age=int(input('Enter Your age:'))
height=float(input('Enter Your Height:'))
is_student=input('Are you a student?(yes/no):').lower()=="yes"
birth_year=2025-age

print('\n'+'='*40, '\n          Personal Summary', '\n'+'='*40)          #Display Personal Summary
print(f'\nFull Name: {full_name}')
print('Age:', age)
print(f'Height: {height} Feet')
print('Birth Year:',birth_year)
print(f"Student Status: {'Yes' if is_student else 'No'}")
print('\n'+'='*40)
```
