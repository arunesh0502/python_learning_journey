```python
# University Comparison using Dictionaries

# Dictionaries for each university
uwa = {
    "name": "University of Western Australia",
    "city": "Perth",
    "tuition": 90000,
    "duration_months": 24,
    "ranking": 77,
    "living_cost_monthly": 1800
}

melbourne = {
    "name": "University of Melbourne",
    "city": "Melbourne",
    "tuition": 120000,
    "duration_months": 24,
    "ranking": 19,
    "living_cost_monthly": 2400
}

monash = {
    "name": "Monash University",
    "city": "Melbourne",
    "tuition": 104000,
    "duration_months": 18,
    "ranking": 36,
    "living_cost_monthly": 2400
}

sydney = {
    "name": "University of Sydney",
    "city": "Sydney",
    "tuition": 116000,
    "duration_months": 24,
    "ranking": 25,
    "living_cost_monthly": 2400
}

# Calculate total costs
for uni in [uwa, melbourne, monash, sydney]:
    total_living = uni["living_cost_monthly"] * uni["duration_months"]
    total_cost = uni["tuition"] + total_living
    
    print(f"\n{uni['name']}")
    print(f"  City              : {uni['city']}")
    print(f"  Tuition           : ${uni['tuition']:,}")
    print(f"  Living Cost ({uni['duration_months']} months): ${total_living:,}")
    print(f"  Total Program Cost: ${total_cost:,}")
    print(f"  QS Ranking        : #{uni['ranking']}")

# Find cheapest option
universities = [uwa, melbourne, monash, sydney]
cheapest = min(universities, key=lambda x: x["tuition"] + (x["living_cost_monthly"] * x["duration_months"]))

print(f"\n✓ Best Value: {cheapest['name']} in {cheapest['city']}")
```
