```python
print("="*40)
print("     WELCOME TO PYTHON PIZZA PALACE")
print("="*40 + "\n")

# Customer information
customer_name = input("Enter your name: ")
phone_number = input("Enter your phone number: ")

print(f"\nHello {customer_name}! Let's take your order.\n")

# Menu items and prices
pizza_price = 12.99
drink_price = 2.50
dessert_price = 4.99
tax_rate = 0.08  # 8% tax

# Order input
num_pizzas = int(input("How many pizzas would you like? "))
num_drinks = int(input("How many drinks would you like? "))
num_desserts = int(input("How many desserts would you like? "))

# Calculations
pizza_subtotal = num_pizzas * pizza_price
drink_subtotal = num_drinks * drink_price
dessert_subtotal = num_desserts * dessert_price
subtotal = pizza_subtotal + drink_subtotal + dessert_subtotal
tax_amount = subtotal * tax_rate
total = subtotal + tax_amount

# Receipt
print("\n" + "="*45)
print("              ORDER RECEIPT")
print("="*45)
print(f"Customer: {customer_name}")
print(f"Phone: {phone_number}")
print("-"*45)
print(f"Pizzas    {num_pizzas} × ${pizza_price:.2f} = ${pizza_subtotal:.2f}")
print(f"Drinks    {num_drinks} × ${drink_price:.2f} = ${drink_subtotal:.2f}")
print(f"Desserts  {num_desserts} × ${dessert_price:.2f} = ${dessert_subtotal:.2f}")
print("-"*45)
print(f"Subtotal:                    ${subtotal:.2f}")
print(f"Tax ({tax_rate:.0%}):                     ${tax_amount:.2f}")
print("="*45)
print(f"TOTAL:                       ${total:.2f}")
print("="*45)
print("Thank you for your order! \n")
```
