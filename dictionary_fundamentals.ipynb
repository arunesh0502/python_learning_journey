```python
# Basic syntax: curly braces { }
# Format: {key: value, key: value}

# Student information
student = {
    "name": "Arun",
    "age": 25,
    "city": "Chennai",
    "moving_to": "Perth",
    "course": "Business Analytics"
}

# University information
university = {
    "name": "UWA",
    "location": "Perth",
    "ranking": 85,
    "tuition": 48000
}

# Empty dictionary
empty_dict = {}
```

```python
# Accessing Dictionary
# Using square bracket method [] 
name = student['name']
age = student['age']
print(name)
print(age)

# country = student['country'] will return error

# Using .get() method - Safer access
name = university.get('name')
ranking = university.get('ranking')
print(name)
print(ranking)

# country = university.get('country') will return 'none'
country = university.get('country','Australia') # add value to the key
print(country)
```

```python
# Modifying Dictionary
# Modifying existing value
student['name'] = 'Arunesh'
university['ranking'] = 77
print(name)
print(ranking)

# Adding new key and value
student['dob'] = '05-02-2001'

# Deleting a key-value pair
del student['moving_to']

# Remove and return value
course = student.pop('course')

print(student)
```
